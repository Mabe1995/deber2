### Medidas de tendencia central y dispersión de la felicidad (data)

En esta sección se calculará una de las medidas de tendencia central y dispersión de la felicidad a nivel mundial para el primer y último año incluido en el set de datos.

```python
PrimerAño = data['anio'].min()
PrimerAño
```

```python
data_2005 = data[data['anio'] == PrimerAño]
data_2005
```

Vamos a tomar la variable 'escalera_vida' en este caso es la variable felicidad. La medida que se interpretará es la media ya que así podemos conocer en promedio que tan feliz esta un país en la escala del 0 al 10. Mientras más alto el puntaje, mayor felicidad.

```python
media_2005= data_2005['escalera_vida'].mean()
media_2005 
```

Una media de felicidad de 6.446 en una escala de 0 a 10 sugiere que, en promedio, las personas en el grupo o población que se está evaluando tienen un nivel de felicidad ligeramente por encima del punto medio de la escala, podríamos decir que es una media moderada de felicidad, a continuación compararemos con el último año para saber si la media de la felicidad aumenta o disminuye 

```python
UltimoAño = data['anio'].max()
UltimoAño
```

```python
data_2018 = data[data['anio'] == UltimoAño]
data_2018
```

```python
media_2018= data_2018['escalera_vida'].mean()
media_2018 
```

En comparación con el primer año, podemos ver que en promedio la felicidad disminuyó y eso puede deberser a diferentes aspectos emocionales, sociales, económicos, políticos, etc. Que afectan de una a otra manera a la población.


#### Medida de dispersión

De la misma manera calcularemos la medida de dispersión, en este caso la desviación estándar del primer y último año

```python
ds_2005= data_2005['escalera_vida'].std()
ds_2005 
```

Una desviación estándar de 0.919 en una escala de 0 a 10 sugiere que los valores de felicidad tienden a estar relativamente cerca de la media, lo que indica una baja dispersión o variabilidad en los datos. Una baja desviación estándar sugiere que las puntuaciones de felicidad están agrupadas más cerca de la media. En otras palabras, la mayoría de las personas en el grupo o población tienen niveles de felicidad que son bastante similares entre sí, lo que indica una mayor consistencia en los niveles de felicidad dentro de la población. Ahora, compararemos con la desviación estándar en el último año.

```python
ds_2018= data_2018['escalera_vida'].std()
ds_2018 
```

En el último año podemos ver que, la desviación estándar aumentó y esto indica que los niveles de felicidad se han vuelto más diversos y que hay una mayor variabilidad en las puntuaciones de felicidad en comparación con el primer año.

Este aumento en la desviación estándar podría deberse a una serie de factores, como cambios en las circunstancias económicas, sociales o personales que hayan afectado a la población en ese período de tiempo. En términos generales, una desviación estándar más alta indica una mayor dispersión de datos alrededor de la media, lo que sugiere que los niveles de felicidad en la población se han vuelto más heterogéneos en el último año.
